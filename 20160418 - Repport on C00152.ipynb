```python
mat, d1, a1 = process(subs, meth='lin', kind='bin', vert_name="DiffLinBin", hor_name='AbilLinBin', verbose=False)
mat, d2, a2 = process(subs, meth='lin', kind='means', vert_name="DiffLinMeans", hor_name='AbilLinMeans', verbose=False)
mat, d3, a3 = process(subs, meth='svd', kind='bin', vert_name="DiffSvdBin", hor_name='AbilSvdBin', verbose=False)
mat, d4, a4 = process(subs, meth='svd', kind='means', vert_name="DiffSvdMeans", hor_name='AbilSvdMeans', verbose=False)
```
