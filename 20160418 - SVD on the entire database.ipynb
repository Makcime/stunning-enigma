```python
%matplotlib inline
%matplotlib notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from analyser import *
```

```python
subs = import_all_submissions(dbname="test1")
#subs = import_submissions(course_id = "C00198", dbname="test1")
```

```python
from timeit import default_timer as timer
start = timer()
a = bianary_mat(subs)
b, c = tots_mats(subs)
d = b / c
d.fillna(0, inplace=True)
end = timer()
print(end - start), 'seconds'
```

```python
npmat = d.as_matrix()
U, s, V = np.linalg.svd(npmat, full_matrices=False)
```

```python
U = pd.DataFrame(U, index=d.index)
SvdDiff = pd.Series(U[0], index=d.index, name="Diff_svd")
SvdDiff = SvdDiff.apply(abs)
```

```python
SvdAbil = pd.Series(V[0], index=d.columns, name="Abil_svd")
SvdAbil = SvdAbil.apply(abs)
```
